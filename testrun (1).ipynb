```python
# =========================================================
#  EfficientNetB3 fundus classifier (DR + Glaucoma + Cataract + Normal)
#  - Kaggle-friendly path detection (recursive, underscore/space tolerant)
#  - Optional offline EfficientNet weights (local file) or ImageNet (Internet ON)
#  - 70/20/10 stratified split
#  - 300x300 + light augmentations
#  - AdamW + CosineDecay (no restarts) + label smoothing + grad clipping
#  - Per-epoch CSV logs (train_history + medical metrics)
#  - Robust checkpoints + resume + best .h5 save
#  - Full evaluation + bootstrap CIs + Grad-CAM (no layer-name errors)
# =========================================================

# !pip install -q "tensorflow==2.14.*" matplotlib scikit-learn pandas numpy opencv-python

import os, glob, math, random, pathlib, json, time, csv, warnings
warnings.filterwarnings("ignore")

os.environ["TF_DETERMINISTIC_OPS"] = "1"
os.environ["TF_CUDNN_DETERMINISTIC"] = "1"

import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.applications import EfficientNetB3

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    classification_report, confusion_matrix, matthews_corrcoef,
    f1_score, jaccard_score, precision_score, recall_score, roc_auc_score
)
import matplotlib.pyplot as plt
import cv2

# ----------------------------
# 0) Reproducibility (best-effort)
# ----------------------------
SEED = 1234
os.environ["PYTHONHASHSEED"] = str(SEED)
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)

# ----------------------------
# 1) Dataset path detection (robust & recursive)
# ----------------------------
from pathlib import Path
from collections import deque

ROOT_CANDIDATES = [
    "/kaggle/input/eye-diseases-classification",
    "/kaggle/input/eye-diseases-classification/dataset",
    "/kaggle/input"
]

def _norm(s: str) -> str:
    return s.lower().replace("_", " ").strip()

EXPECTED = {"cataract", "diabetic retinopathy", "glaucoma", "normal"}

def find_data_dir_recursive(roots, max_depth=4):
    for root in roots:
        rootp = Path(root)
        if not rootp.exists():
            continue
        dq = deque([(rootp, 0)])
        while dq:
            cur, depth = dq.popleft()
            if depth > max_depth:
                continue
            if cur.is_dir():
                children = [c.name for c in cur.iterdir() if c.is_dir()]
                if { _norm(c) for c in children } >= EXPECTED:
                    return str(cur), sorted(children)
                for child in cur.iterdir():
                    if child.is_dir():
                        dq.append((child, depth + 1))
    raise FileNotFoundError(
        f"Could not find class folders {EXPECTED} under roots: {roots}. "
        f"Run `!ls -R /kaggle/input | head -n 200` to inspect if needed."
    )

DATA_DIR, classes = find_data_dir_recursive(ROOT_CANDIDATES, max_depth=4)
print("Using DATA_DIR:", DATA_DIR)
print("Found classes:", classes)

display_names = [c.replace("_"," ") for c in classes]
class_to_id = {c:i for i,c in enumerate(classes)}
id_to_class = {i:c for c,i in class_to_id.items()}
num_classes = len(classes)
print("Label mapping:", class_to_id)

# ----------------------------
# 2) File list & 70/20/10 split (stratified) + save splits
# ----------------------------
def list_images(root):
    fps, labs = [], []
    for cls in classes:
        for fp in glob.glob(os.path.join(root, cls, "*")):
            if fp.lower().endswith((".jpg",".jpeg",".png",".bmp",".tif",".tiff")):
                fps.append(fp); labs.append(cls)
    return pd.DataFrame({"filepath": fps, "label": labs})

all_df = list_images(DATA_DIR)
print("Total images:", len(all_df))
print(all_df.groupby("label").size())

y_all = all_df["label"].map(class_to_id).values
train_df, temp_df = train_test_split(
    all_df, test_size=0.30, random_state=SEED, stratify=y_all
)
y_temp = temp_df["label"].map(class_to_id).values
val_df, test_df = train_test_split(
    temp_df, test_size=1/3, random_state=SEED, stratify=y_temp
)

print("Split sizes -> Train:", len(train_df), " Val:", len(val_df), " Test:", len(test_df))
print("Train dist:\n", train_df.groupby("label").size())
print("Val   dist:\n", val_df.groupby("label").size())
print("Test  dist:\n", test_df.groupby("label").size())

os.makedirs("splits", exist_ok=True)
train_df.to_csv("splits/train.csv", index=False)
val_df.to_csv("splits/val.csv", index=False)
test_df.to_csv("splits/test.csv", index=False)

# ----------------------------
# 3) TF Dataset pipeline (300x300, normalization, augmentations)
# ----------------------------
IMG_SIZE = 300   # EfficientNetB3 canonical size (helps metrics)
BATCH_SIZE = 32
AUTOTUNE = tf.data.AUTOTUNE

augment = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.10),       # ~±18°
    layers.RandomZoom(0.15, 0.15),
    layers.RandomContrast(0.15),
], name="augmentation")

def load_image(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=3, expand_animations=False)
    img = tf.image.convert_image_dtype(img, tf.float32)  # [0,1]
    img = tf.image.resize(img, (IMG_SIZE, IMG_SIZE), antialias=True)
    return img, tf.one_hot(label, num_classes)

def make_ds(df, training=False):
    paths = df["filepath"].values
    y = df["label"].map(class_to_id).values
    ds = tf.data.Dataset.from_tensor_slices((paths, y))
    if training:
        ds = ds.shuffle(buffer_size=len(df), seed=SEED, reshuffle_each_iteration=True)
    ds = ds.map(lambda p,lab: load_image(p, lab), num_parallel_calls=AUTOTUNE)
    if training:
        ds = ds.map(lambda x,y: (augment(x, training=True), y), num_parallel_calls=AUTOTUNE)
    return ds.batch(BATCH_SIZE).prefetch(AUTOTUNE)

train_ds = make_ds(train_df, training=True)
val_ds   = make_ds(val_df,   training=False)
test_ds  = make_ds(test_df,  training=False)

# ----------------------------
# 4) EfficientNetB3 base (offline-friendly weights) + HEAD (no BN bug)
# ----------------------------
LOCAL_WEIGHTS = "/kaggle/input/efficientnetb3-weights/efficientnetb3_notop.h5"  # optional

def load_efficientnet_b3(input_shape=(IMG_SIZE, IMG_SIZE, 3)):
    try:
        if os.path.exists(LOCAL_WEIGHTS):
            print("Loading EfficientNetB3 weights from local file:", LOCAL_WEIGHTS)
            base = EfficientNetB3(include_top=False, weights=None, input_shape=input_shape)
            base.load_weights(LOCAL_WEIGHTS)
            return base
        else:
            print("Local weights not found. Trying ImageNet (requires Internet ON).")
            return EfficientNetB3(include_top=False, weights="imagenet", input_shape=input_shape)
    except Exception as e:
        print("Weights load failed, falling back to random init:", e)
        return EfficientNetB3(include_top=False, weights=None, input_shape=input_shape)

keras.backend.clear_session()
base = load_efficientnet_b3()
base.trainable = True  # we'll warm up frozen, then unfreeze

inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
x = base(inputs)  # DO NOT set training=True here
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.3)(x)
x = layers.Dense(256, activation="relu")(x)
x = layers.Dropout(0.3)(x)
outputs = layers.Dense(num_classes, activation="softmax")(x)
model = keras.Model(inputs, outputs)
model.summary()

# ----------------------------
# 5) Optimizer & schedules + callbacks (incl. CSV logging & checkpoints)
# ----------------------------
TOTAL_EPOCHS   = 70
WARMUP_EPOCHS  = 3
PHASE2_EPOCHS  = TOTAL_EPOCHS - WARMUP_EPOCHS

# Cosine decay WITHOUT restarts (smoother validation)
steps_per_epoch = max(1, math.ceil(len(train_df) / BATCH_SIZE))
cosine = keras.optimizers.schedules.CosineDecay(
    initial_learning_rate=1e-4,
    decay_steps=steps_per_epoch * PHASE2_EPOCHS,
    alpha=1e-6
)

# Use AdamW with mild weight decay + grad clipping (helps generalization)
def make_optimizer(lr_schedule_or_float):
    return keras.optimizers.AdamW(
        learning_rate=lr_schedule_or_float,
        weight_decay=1e-5,
        clipnorm=1.0
    )

# Label smoothing (reduces overconfidence; often boosts medical F1)
loss_fn = keras.losses.CategoricalCrossentropy(label_smoothing=0.05)

os.makedirs("logs", exist_ok=True)
os.makedirs("ckpts", exist_ok=True)
os.makedirs("backup", exist_ok=True)

csv_logger = keras.callbacks.CSVLogger("logs/train_history.csv", separator=",", append=True)

# Per-epoch clinical metrics -> logs/val_medical_metrics.csv
class MedMetricsCSV(keras.callbacks.Callback):
    def __init__(self, val_ds, class_names, csv_path="logs/val_medical_metrics.csv"):
        super().__init__()
        self.val_ds = val_ds
        self.class_names = list(class_names)
        self.csv_path = csv_path
        norm = lambda s: s.lower().replace("_"," ").strip()
        name_to_idx = {norm(c): i for i, c in enumerate(self.class_names)}
        self.idx_dr = name_to_idx.get("diabetic retinopathy", None)
        self.idx_gl = name_to_idx.get("glaucoma", None)
        if not os.path.exists(self.csv_path):
            with open(self.csv_path, "w", newline="") as f:
                cols = [
                    "epoch","lr","val_loss","val_accuracy",
                    "macro_precision","macro_recall","macro_specificity",
                    "macro_f1","macro_jaccard","mcc","macro_auroc",
                    "sens_dr","spec_dr","sens_glaucoma","spec_glaucoma"
                ]
                csv.writer(f).writerow(cols)

    def _current_lr(self):
        try:
            lr = self.model.optimizer.learning_rate
            if isinstance(lr, tf.keras.optimizers.schedules.LearningRateSchedule):
                return float(lr(self.model.optimizer.iterations).numpy())
            return float(tf.keras.backend.get_value(lr))
        except Exception:
            return float("nan")

    def on_epoch_end(self, epoch, logs=None):
        logs = logs or {}
        y_true, y_prob = [], []
        for bx, by in self.val_ds:
            y_true.append(np.argmax(by.numpy(), axis=1))
            y_prob.append(self.model.predict(bx, verbose=0))
        y_true = np.concatenate(y_true); y_prob = np.concatenate(y_prob)
        y_pred = np.argmax(y_prob, axis=1)

        macro_prec = precision_score(y_true, y_pred, average="macro", zero_division=0)
        macro_rec  = recall_score(y_true, y_pred, average="macro", zero_division=0)
        macro_f1   = f1_score(y_true, y_pred, average="macro", zero_division=0)
        macro_jacc = jaccard_score(y_true, y_pred, average="macro", zero_division=0)
        mcc        = matthews_corrcoef(y_true, y_pred)

        labels = np.arange(len(self.class_names))
        cm = confusion_matrix(y_true, y_pred, labels=labels)
        sens = np.diag(cm) / np.maximum(1, cm.sum(axis=1))
        total = cm.sum(); row_sum = cm.sum(axis=1); col_sum = cm.sum(axis=0)
        tn = total - (row_sum + col_sum - np.diag(cm))
        fp = col_sum - np.diag(cm)
        spec = tn / np.maximum(1, tn + fp)
        macro_spec = np.nanmean(spec)

        try:
            macro_auroc = roc_auc_score(y_true, y_prob, multi_class="ovr", average="macro")
        except Exception:
            macro_auroc = np.nan

        sens_dr = float(sens[self.idx_dr]) if self.idx_dr is not None else np.nan
        spec_dr = float(spec[self.idx_dr]) if self.idx_dr is not None else np.nan
        sens_gl = float(sens[self.idx_gl]) if self.idx_gl is not None else np.nan
        spec_gl = float(spec[self.idx_gl]) if self.idx_gl is not None else np.nan

        row = {
            "epoch": int(epoch+1),
            "lr": self._current_lr(),
            "val_loss": float(logs.get("val_loss", np.nan)),
            "val_accuracy": float(logs.get("val_accuracy", np.nan)),
            "macro_precision": float(macro_prec),
            "macro_recall": float(macro_rec),
            "macro_specificity": float(macro_spec),
            "macro_f1": float(macro_f1),
            "macro_jaccard": float(macro_jacc),
            "mcc": float(mcc),
            "macro_auroc": float(macro_auroc),
            "sens_dr": sens_dr, "spec_dr": spec_dr,
            "sens_glaucoma": sens_gl, "spec_glaucoma": spec_gl
        }
        with open(self.csv_path, "a", newline="") as f:
            csv.DictWriter(f, fieldnames=row.keys()).writerow(row)

        print(
            f"[epoch {row['epoch']}] val_acc={row['val_accuracy']:.4f} "
            f"F1(macro)={row['macro_f1']:.4f} Sens(macro)={row['macro_recall']:.4f} "
            f"Spec(macro)={row['macro_specificity']:.4f} MCC={row['mcc']:.4f} "
            f"AUROC(macro)={row['macro_auroc']:.4f} | "
            f"Sens_DR={row['sens_dr']:.4f} Spec_DR={row['spec_dr']:.4f} "
            f"Sens_GLAU={row['sens_glaucoma']:.4f} Spec_GLAU={row['spec_glaucoma']:.4f}"
        )

med_logger = MedMetricsCSV(val_ds=val_ds, class_names=classes, csv_path="logs/val_medical_metrics.csv")

callbacks = [
    # No EarlyStopping -> will run full 70 epochs
    keras.callbacks.ModelCheckpoint(
        filepath="efnb3_best.keras",
        monitor="val_accuracy",
        save_best_only=True,
        save_weights_only=False
    ),
    keras.callbacks.ModelCheckpoint(
        filepath="efnb3_best.h5",
        monitor="val_accuracy",
        save_best_only=True,
        save_weights_only=False
    ),
    keras.callbacks.ModelCheckpoint(
        filepath="ckpts/latest.weights.h5",
        monitor="val_accuracy",
        save_best_only=False,
        save_weights_only=True
    ),
    keras.callbacks.BackupAndRestore(backup_dir="backup"),
    csv_logger,
    med_logger
]

# Optional resume
resume_path = "ckpts/latest.weights.h5"
if os.path.exists(resume_path):
    try:
        model.load_weights(resume_path)
        print(f"Resumed model weights from: {resume_path}")
    except Exception as e:
        print(f"Could not load {resume_path}: {e}")

# ----------------------------
# 6) Train (2-phase: warm-up frozen, then fine-tune)
# ----------------------------
# Phase 1: warm-up (freeze backbone)
base.trainable = False
model.compile(optimizer=make_optimizer(3e-4), loss=loss_fn, metrics=["accuracy"])
history1 = model.fit(train_ds, validation_data=val_ds, epochs=WARMUP_EPOCHS, callbacks=callbacks, verbose=1)

# Phase 2: unfreeze & fine-tune
base.trainable = True
model.compile(optimizer=make_optimizer(cosine), loss=loss_fn, metrics=["accuracy"])
history2 = model.fit(train_ds, validation_data=val_ds, epochs=PHASE2_EPOCHS, callbacks=callbacks, verbose=1)

# Merge histories for plotting
history = type("H", (), {})()
history.history = {}
for k in set(list(history1.history.keys()) + list(history2.history.keys())):
    history.history[k] = history1.history.get(k, []) + history2.history.get(k, [])

# ----------------------------
# 7) Plot training curves
# ----------------------------
def plot_history(h):
    fig, ax = plt.subplots(1,2, figsize=(12,4))
    ax[0].plot(h.history.get("accuracy", [])); ax[0].plot(h.history.get("val_accuracy", []))
    ax[0].set_title("Accuracy"); ax[0].legend(["train","val"])
    ax[1].plot(h.history.get("loss", [])); ax[1].plot(h.history.get("val_loss", []))
    ax[1].set_title("Loss"); ax[1].legend(["train","val"])
    plt.tight_layout(); plt.show()
plot_history(history)

# ----------------------------
# 8) Final evaluation on test set
# ----------------------------
y_true, y_prob = [], []
for bx, by in test_ds:
    y_true.append(np.argmax(by.numpy(), axis=1))
    y_prob.append(model.predict(bx, verbose=0))
y_true = np.concatenate(y_true); y_prob = np.concatenate(y_prob)
y_pred = np.argmax(y_prob, axis=1)

test_acc = (y_true == y_pred).mean()
print(f"\nTEST Accuracy: {test_acc*100:.2f}%")

print("\nTEST Classification Report:")
print(classification_report(y_true, y_pred, target_names=display_names, digits=4))

cm = confusion_matrix(y_true, y_pred, labels=np.arange(num_classes))
print("\nTEST Confusion Matrix:\n", cm)

mcc = matthews_corrcoef(y_true, y_pred)
dice_macro = f1_score(y_true, y_pred, average="macro")
dice_weighted = f1_score(y_true, y_pred, average="weighted")
jaccard_macro = jaccard_score(y_true, y_pred, average="macro")
jaccard_weighted = jaccard_score(y_true, y_pred, average="weighted")
try:
    macro_auroc_test = roc_auc_score(y_true, y_prob, multi_class="ovr", average="macro")
except Exception:
    macro_auroc_test = np.nan

print(f"\nMCC: {mcc:.4f}")
print(f"Dice (macro): {dice_macro:.4f}   Dice (weighted): {dice_weighted:.4f}")
print(f"Jaccard (macro): {jaccard_macro:.4f}   Jaccard (weighted): {jaccard_weighted:.4f}")
print(f"AUROC (macro): {macro_auroc_test:.4f}")

def plot_cm(cm, labels, title="Confusion Matrix"):
    fig, ax = plt.subplots(figsize=(6,5))
    im = ax.imshow(cm, cmap="Blues")
    ax.set_xticks(range(len(labels))); ax.set_yticks(range(len(labels)))
    ax.set_xticklabels(labels, rotation=45, ha="right"); ax.set_yticklabels(labels)
    for (i,j), v in np.ndenumerate(cm):
        ax.text(j, i, str(v), ha="center", va="center")
    ax.set_xlabel("Predicted"); ax.set_ylabel("True"); ax.set_title(title)
    plt.colorbar(im); plt.tight_layout(); plt.show()
plot_cm(cm, display_names)

# ----------------------------
# 9) Bootstrap 95% CIs for key metrics
# ----------------------------
def bootstrap_ci(metric_fn, y_true, y_pred, n=1000, seed=SEED):
    rng = np.random.default_rng(seed)
    idx = np.arange(len(y_true)); vals = []
    for _ in range(n):
        b = rng.choice(idx, size=len(idx), replace=True)
        vals.append(metric_fn(y_true[b], y_pred[b]))
    vals = np.array(vals)
    lo, hi = np.percentile(vals, [2.5, 97.5])
    return float(vals.mean()), (float(lo), float(hi))

acc_mean, acc_ci = bootstrap_ci(lambda yt, yp: (yt == yp).mean(), y_true, y_pred)
f1_mean, f1_ci   = bootstrap_ci(lambda yt, yp: f1_score(yt, yp, average="macro"), y_true, y_pred)
mcc_mean, mcc_ci = bootstrap_ci(lambda yt, yp: matthews_corrcoef(yt, yp), y_true, y_pred)

print("\nBootstrapped 95% CI:")
print(f" Accuracy mean={acc_mean:.4f}  CI=({acc_ci[0]:.4f}, {acc_ci[1]:.4f})")
print(f" F1-macro mean={f1_mean:.4f}  CI=({f1_ci[0]:.4f}, {f1_ci[1]:.4f})")
print(f" MCC mean={mcc_mean:.4f}      CI=({mcc_ci[0]:.4f}, {mcc_ci[1]:.4f})")

# ----------------------------
# 10) Grad-CAM (robust for nested EfficientNet — FIXES your error)
# ----------------------------
from tensorflow.keras import Model

# Get the EfficientNet submodel actually inside your model
eff_submodel = model.get_layer('efficientnetb3')

# Try canonical last conv; else fallback to last Conv2D in the submodel
try:
    last_conv_layer = eff_submodel.get_layer('top_conv')
except Exception:
    last_conv_layer = next(l for l in reversed(eff_submodel.layers) if isinstance(l, layers.Conv2D))

print("Grad-CAM using last conv layer:", last_conv_layer.name)

def grad_cam(img_tensor, target_class_idx=None, conv_layer=None):
    if conv_layer is None: conv_layer = last_conv_layer
    conv_tensor = conv_layer.output  # layer OBJECT, not name string

    grad_model = keras.models.Model(
        inputs=model.inputs,
        outputs=[conv_tensor, model.output]
    )

    with tf.GradientTape() as tape:
        conv_outputs, preds = grad_model(img_tensor, training=False)
        if target_class_idx is None:
            target_class_idx = tf.argmax(preds[0])
        loss = preds[:, target_class_idx]

    grads = tape.gradient(loss, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0,1,2))
    conv_outputs = conv_outputs[0]
    heatmap = tf.reduce_sum(tf.multiply(pooled_grads, conv_outputs), axis=-1)
    heatmap = tf.maximum(heatmap, 0) / (tf.reduce_max(heatmap) + 1e-9)
    return heatmap.numpy()

# Example visualization
sample = test_df.sample(1, random_state=SEED)["filepath"].iloc[0]
raw = tf.io.read_file(sample)
raw = tf.image.decode_image(raw, channels=3, expand_animations=False)
rawf = tf.image.convert_image_dtype(raw, tf.float32)
inp = tf.image.resize(rawf, (IMG_SIZE, IMG_SIZE))[None, ...]
pred = model.predict(inp, verbose=0)[0]
pred_idx = int(np.argmax(pred))
heat = grad_cam(inp, pred_idx)

h = cv2.resize(heat, (IMG_SIZE, IMG_SIZE))
h = (255*h).astype("uint8")
h = cv2.applyColorMap(h, cv2.COLORMAP_JET)[:, :, ::-1] / 255.0
overlay = 0.35*h + 0.65*rawf.numpy()

plt.figure(figsize=(9,4))
plt.subplot(1,3,1); plt.imshow(rawf.numpy()); plt.title(f"True: {pathlib.Path(sample).parent.name}")
plt.subplot(1,3,2); plt.imshow(overlay); plt.title(f"Grad-CAM → {display_names[pred_idx]}")
plt.subplot(1,3,3); plt.bar(range(num_classes), pred); plt.xticks(range(num_classes), display_names, rotation=45)
plt.tight_layout(); plt.show()

# ----------------------------
# 11) Save artifacts (best are already saved by callbacks)
# ----------------------------
try:
    model.save("efnb3_final.keras")
    model.save("efnb3_final.h5")
except Exception as e:
    print("Final HDF5 save failed; saving weights-only fallback:", e)
    model.save_weights("efnb3_final.weights.h5")

print("\nArtifacts written:")
print(" - Best models:  efnb3_best.keras , efnb3_best.h5")
print(" - Last models:  efnb3_final.keras , efnb3_final.h5 (or .weights.h5)")
print(" - Latest ckpt:  ckpts/latest.weights.h5  (for resume)")
print(" - Splits:       splits/train.csv , splits/val.csv , splits/test.csv")
print(" - Logs:         logs/train_history.csv , logs/val_medical_metrics.csv")
```

```python
# ======================================================
# Grad-CAM for your saved EfficientNetB3 model (no retrain)
# - Avoids sub-model graph KeyErrors (single-tape forward)
# - Uses final conv activation from EfficientNet base
# - Safe overlay sizing (no broadcasting mismatch)
# ======================================================

import os, random, pathlib
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import cv2
import matplotlib.pyplot as plt

# (Optional) makes things simpler during one-off inference
tf.config.run_functions_eagerly(True)

# --------- CONFIG ---------
MODEL_CANDIDATES = [
    "efnb3_best.keras",
    "efnb3_final.keras",
    "efnb3_best.h5",
    "efnb3_final.h5",
]
# Set this to force a specific image; otherwise the script will auto-pick one.
IMG_PATH = None  # e.g. "/kaggle/input/eye-diseases-classification/dataset/glaucoma/_4_5317326.jpg"
CLASS_NAMES = ["cataract", "diabetic retinopathy", "glaucoma", "normal"]
DATA_ROOTS = [
    "/kaggle/input/eye-diseases-classification/dataset",
    "/kaggle/input/eye-diseases-classification",
    "/kaggle/input"
]
# --------------------------

# 1) Load your saved model
model = None
for p in MODEL_CANDIDATES:
    if os.path.exists(p):
        model = keras.models.load_model(p)
        print(f"Loaded saved model: {p}")
        break
if model is None:
    raise FileNotFoundError("No saved model found among: " + ", ".join(MODEL_CANDIDATES))

# Infer model input size (your logs show 300x300x3)
_, H, W, C = model.input_shape
assert C == 3, f"Expected 3-channel input, got {C}"
IMG_SIZE = (H, W)

# Grab the EfficientNet base inside your model (name from your summary)
base = model.get_layer("efficientnetb3")
# The base output is the last activation map (10x10x1536 for B3)
try:
    last_act_name = base.get_layer("top_activation").name
except Exception:
    last_act_name = "base_output"
print("Grad-CAM uses base activation:", last_act_name)

# 2) Pick an image
def auto_pick_image():
    exts = (".jpg",".jpeg",".png",".bmp",".tif",".tiff")
    # Prefer class subfolders (keeps things realistic)
    for root in DATA_ROOTS:
        if not os.path.exists(root): 
            continue
        candidates = []
        for cls in ["cataract", "diabetic_retinopathy", "glaucoma", "normal"]:
            d = os.path.join(root, cls)
            if os.path.isdir(d):
                for p in pathlib.Path(d).glob("*"):
                    sp = str(p).lower()
                    if sp.endswith(exts):
                        candidates.append(str(p))
        if candidates:
            random.shuffle(candidates)
            return candidates[0]
    # Fallback: any image under roots
    for root in DATA_ROOTS:
        if not os.path.exists(root): 
            continue
        any_imgs = [str(p) for p in pathlib.Path(root).rglob("*") if str(p).lower().endswith(exts)]
        if any_imgs:
            random.shuffle(any_imgs)
            return any_imgs[0]
    raise FileNotFoundError("Could not find any image. Set IMG_PATH manually.")

if IMG_PATH is None:
    IMG_PATH = auto_pick_image()
print("Using image:", IMG_PATH)

# 3) Load & preprocess
def load_for_model(path):
    raw = tf.io.read_file(path)
    img = tf.image.decode_image(raw, channels=3, expand_animations=False)
    img = tf.image.convert_image_dtype(img, tf.float32)  # [0,1]
    orig = img.numpy()  # keep original for overlay (H0,W0,3)
    x = tf.image.resize(img, IMG_SIZE, antialias=True)[None, ...]  # (1,H,W,3)
    return orig, x

orig_img, x = load_for_model(IMG_PATH)

# 4) Build the SAME head as in your trained model (reuse layers by name)
#    From your summary: global_average_pooling2d -> dropout -> dense -> dropout_1 -> dense_1
gap_layer   = model.get_layer("global_average_pooling2d")
drop1_layer = model.get_layer("dropout")
fc_layer    = model.get_layer("dense")
drop2_layer = model.get_layer("dropout_1")
out_layer   = model.get_layer("dense_1")

# 5) Grad-CAM in a single tape (no sub-model construction = no KeyError)
@tf.function(jit_compile=False)  # stay simple/robust
def forward_with_cam(inp):
    with tf.GradientTape() as tape:
        # features from EfficientNet base (shape ~ [1,10,10,1536])
        feats = base(inp, training=False)
        tape.watch(feats)  # ensure grads wrt features

        # head forward
        z = gap_layer(feats)
        z = drop1_layer(z, training=False)
        z = fc_layer(z)
        z = drop2_layer(z, training=False)
        preds = out_layer(z)  # shape [1,4], softmax from training (kept)

        # pick class idx (top1)
        cls_idx = tf.argmax(preds[0])
        loss = preds[:, cls_idx]

    # gradient of selected prob wrt features
    grads = tape.gradient(loss, feats)              # [1,h,w,c]
    weights = tf.reduce_mean(grads, axis=(1,2))     # [1,c]
    # weighted sum over channels
    cam = tf.reduce_sum(feats * weights[:, None, None, :], axis=-1)  # [1,h,w]
    cam = tf.nn.relu(cam)
    cam = cam[0]
    cam = cam / (tf.reduce_max(cam) + 1e-9)
    return cam, cls_idx, preds[0]

# Warm-up (builds tf.function nicely)
_ = forward_with_cam(tf.zeros_like(x))

# Real run
cam, cls_idx, probs = forward_with_cam(x)
cls_idx = int(cls_idx.numpy())
probs = probs.numpy()

# 6) Overlay (resize CAM to original size to avoid broadcasting issues)
H0, W0 = orig_img.shape[:2]
cam_u8 = (cv2.resize(cam.numpy(), (W0, H0)) * 255).astype("uint8")
heat_color = cv2.applyColorMap(cam_u8, cv2.COLORMAP_JET)[:, :, ::-1] / 255.0  # BGR->RGB
overlay = 0.35*heat_color + 0.65*orig_img
overlay = np.clip(overlay, 0.0, 1.0)

# 7) Show & save
pred_name = CLASS_NAMES[cls_idx] if cls_idx < len(CLASS_NAMES) else f"class_{cls_idx}"
print(f"\nPredicted: {pred_name} (idx={cls_idx})")
print("Probabilities:", np.round(probs, 4))

plt.figure(figsize=(11,4))
plt.subplot(1,3,1); plt.imshow(orig_img);  plt.title("Original"); plt.axis("off")
plt.subplot(1,3,2); plt.imshow(overlay);   plt.title(f"Grad-CAM → {pred_name}"); plt.axis("off")
plt.subplot(1,3,3); plt.bar(range(len(probs)), probs)
plt.xticks(range(len(CLASS_NAMES)), CLASS_NAMES, rotation=45, ha="right"); plt.ylim(0,1)
plt.title("Class probabilities"); plt.tight_layout(); plt.show()

save_name = os.path.splitext(os.path.basename(IMG_PATH))[0] + "_gradcam.png"
cv2.imwrite(save_name, (overlay[:, :, ::-1] * 255).astype("uint8"))
print("Saved overlay to:", save_name)
```

```python
# =========================================================
#  Paper-aligned EfficientNetB3 fundus classifier
#  (end-to-end, Adam + cosine annealing, single hidden head)
# =========================================================

import os, glob, math, random, pathlib, csv, warnings
warnings.filterwarnings("ignore")

# Reproducibility toggles (best-effort)
SEED = 1234
os.environ["PYTHONHASHSEED"] = str(SEED)
random.seed(SEED)
import numpy as np
np.random.seed(SEED)

import tensorflow as tf
tf.random.set_seed(SEED)
os.environ["TF_DETERMINISTIC_OPS"] = "1"
os.environ["TF_CUDNN_DETERMINISTIC"] = "1"

import pandas as pd
import matplotlib.pyplot as plt
import cv2

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.applications import EfficientNetB3
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    classification_report, confusion_matrix, matthews_corrcoef,
    f1_score, jaccard_score, precision_score, recall_score, roc_auc_score
)

# ----------------------------
# Config (strictly paper-aligned)
# ----------------------------
IMG_SIZE      = 300        # train/infer size for the network
BATCH_SIZE    = 32
TOTAL_EPOCHS  = 70
INIT_LR       = 1e-4       # paper: 1e-4–1e-5, start at 1e-4
TRAIN         = True       # set False to skip training and only eval/Grad-CAM

# ----------------------------
# Dataset path detection (kept from your code)
# ----------------------------
from pathlib import Path
from collections import deque

ROOT_CANDIDATES = [
    "/kaggle/input/eye-diseases-classification",
    "/kaggle/input/eye-diseases-classification/dataset",
    "/kaggle/input"
]
def _norm(s: str) -> str: return s.lower().replace("_"," ").strip()
EXPECTED = {"cataract","diabetic retinopathy","glaucoma","normal"}

def find_data_dir_recursive(roots, max_depth=4):
    for root in roots:
        rootp = Path(root)
        if not rootp.exists(): continue
        dq = deque([(rootp, 0)])
        while dq:
            cur, depth = dq.popleft()
            if depth > max_depth: continue
            if cur.is_dir():
                kids = [c.name for c in cur.iterdir() if c.is_dir()]
                if {_norm(k) for k in kids} >= EXPECTED:
                    return str(cur), sorted(kids)
                for child in cur.iterdir():
                    if child.is_dir():
                        dq.append((child, depth+1))
    raise FileNotFoundError("Class folders not found under roots: %s" % roots)

DATA_DIR, classes = find_data_dir_recursive(ROOT_CANDIDATES, max_depth=4)
print("Using DATA_DIR:", DATA_DIR)
print("Found classes:", classes)

display_names = [c.replace("_"," ") for c in classes]
class_to_id = {c:i for i,c in enumerate(classes)}
id_to_class = {i:c for c,i in class_to_id.items()}
num_classes = len(classes)
print("Label mapping:", class_to_id)

# ----------------------------
# File list & 70/20/10 split (same as yours)
# ----------------------------
def list_images(root):
    fps, labs = [], []
    for cls in classes:
        for fp in glob.glob(os.path.join(root, cls, "*")):
            if fp.lower().endswith((".jpg",".jpeg",".png",".bmp",".tif",".tiff")):
                fps.append(fp); labs.append(cls)
    return pd.DataFrame({"filepath": fps, "label": labs})

all_df = list_images(DATA_DIR)
print("Total images:", len(all_df))
print(all_df.groupby("label").size())

y_all = all_df["label"].map(class_to_id).values
train_df, temp_df = train_test_split(all_df, test_size=0.30, random_state=SEED, stratify=y_all)
y_temp = temp_df["label"].map(class_to_id).values
val_df, test_df = train_test_split(temp_df, test_size=1/3, random_state=SEED, stratify=y_temp)

print("Split sizes -> Train:", len(train_df), " Val:", len(val_df), " Test:", len(test_df))
print("Train dist:\n", train_df.groupby("label").size())
print("Val   dist:\n", val_df.groupby("label").size())
print("Test  dist:\n", test_df.groupby("label").size())

os.makedirs("splits", exist_ok=True)
train_df.to_csv("splits/train.csv", index=False)
val_df.to_csv("splits/val.csv", index=False)
test_df.to_csv("splits/test.csv", index=False)

# ----------------------------
# TF Datasets (paper: light aug)
# ----------------------------
AUTOTUNE = tf.data.AUTOTUNE

# Only horizontal flip (conservative augment as in paper)
augment = keras.Sequential([
    layers.RandomFlip("horizontal"),
], name="augmentation")

def load_image(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=3, expand_animations=False)
    img = tf.image.convert_image_dtype(img, tf.float32)          # [0,1]
    img = tf.image.resize(img, (IMG_SIZE, IMG_SIZE), antialias=True)
    return img, tf.one_hot(label, num_classes)

def make_ds(df, training=False):
    paths = df["filepath"].values
    y = df["label"].map(class_to_id).values
    ds = tf.data.Dataset.from_tensor_slices((paths, y))
    if training:
        ds = ds.shuffle(buffer_size=len(df), seed=SEED, reshuffle_each_iteration=True)
    ds = ds.map(lambda p,lab: load_image(p, lab), num_parallel_calls=AUTOTUNE)
    if training:
        ds = ds.map(lambda x,y: (augment(x, training=True), y), num_parallel_calls=AUTOTUNE)
    ds = ds.batch(BATCH_SIZE, drop_remainder=False).prefetch(AUTOTUNE)
    return ds

train_ds = make_ds(train_df, training=True)
val_ds   = make_ds(val_df,   training=False)
test_ds  = make_ds(test_df,  training=False)

# ----------------------------
# Model: EfficientNetB3 + GAP + Dropout + Dense(256, ReLU) + Dropout + Dense(4, Softmax)
# (one hidden layer; Dense layers have bias=True; EfficientNet convs keep bias=False)
# ----------------------------
keras.backend.clear_session()
try:
    base = EfficientNetB3(include_top=False, weights="imagenet", input_shape=(IMG_SIZE, IMG_SIZE, 3))
except Exception:
    print("ImageNet weights unavailable; falling back to random init.")
    base = EfficientNetB3(include_top=False, weights=None, input_shape=(IMG_SIZE, IMG_SIZE, 3))

inputs  = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
x       = base(inputs)
x       = layers.GlobalAveragePooling2D(name="gap")(x)
x       = layers.Dropout(0.3, name="dropout_1")(x)
x       = layers.Dense(256, activation="relu", use_bias=True, name="fc_256")(x)  # single hidden layer
x       = layers.Dropout(0.3, name="dropout_2")(x)
outputs = layers.Dense(num_classes, activation="softmax", use_bias=True, name="pred")(x)
model   = keras.Model(inputs, outputs, name="efnb3_paper_aligned")
model.summary()

# ----------------------------
# Optimizer, Loss, Schedule  (paper: Adam + cosine annealing)
# ----------------------------
# Cosine annealing with restarts across TOTAL_EPOCHS
steps_per_epoch = max(1, math.ceil(len(train_df) / BATCH_SIZE))
T0_steps  = steps_per_epoch * 10     # first cycle ~10 epochs
T_mul     = 2.0                      # each cycle ~doubles
alpha     = 1e-6                     # final LR floor (fraction of init)

lr_schedule = keras.optimizers.schedules.CosineDecayRestarts(
    initial_learning_rate=INIT_LR,
    first_decay_steps=T0_steps,
    t_mul=T_mul,
    m_mul=1.0,
    alpha=alpha
)

optimizer = keras.optimizers.Adam(learning_rate=lr_schedule)
loss_fn   = keras.losses.CategoricalCrossentropy()   # no label smoothing (paper)

# ----------------------------
# Callbacks (same filenames as your code)
# ----------------------------
os.makedirs("logs", exist_ok=True)
os.makedirs("ckpts", exist_ok=True)
os.makedirs("backup", exist_ok=True)

csv_logger = keras.callbacks.CSVLogger("logs/train_history.csv", separator=",", append=True)

class MedMetricsCSV(keras.callbacks.Callback):
    def __init__(self, val_ds, class_names, csv_path="logs/val_medical_metrics.csv"):
        super().__init__()
        self.val_ds, self.class_names = val_ds, list(class_names)
        self.csv_path = csv_path
        norm = lambda s: s.lower().replace("_"," ").strip()
        name_to_idx = {norm(c): i for i, c in enumerate(self.class_names)}
        self.idx_dr = name_to_idx.get("diabetic retinopathy", None)
        self.idx_gl = name_to_idx.get("glaucoma", None)
        if not os.path.exists(self.csv_path):
            with open(self.csv_path, "w", newline="") as f:
                cols = [
                    "epoch","lr","val_loss","val_accuracy",
                    "macro_precision","macro_recall","macro_specificity",
                    "macro_f1","macro_jaccard","mcc","macro_auroc",
                    "sens_dr","spec_dr","sens_glaucoma","spec_glaucoma"
                ]
                csv.writer(f).writerow(cols)
    def _current_lr(self):
        try:
            lr = self.model.optimizer.learning_rate
            if isinstance(lr, tf.keras.optimizers.schedules.LearningRateSchedule):
                return float(lr(self.model.optimizer.iterations).numpy())
            return float(tf.keras.backend.get_value(lr))
        except Exception:
            return float("nan")
    def on_epoch_end(self, epoch, logs=None):
        logs = logs or {}
        y_true, y_prob = [], []
        for bx, by in self.val_ds:
            y_true.append(np.argmax(by.numpy(), axis=1))
            y_prob.append(self.model.predict(bx, verbose=0))
        y_true = np.concatenate(y_true); y_prob = np.concatenate(y_prob)
        y_pred = np.argmax(y_prob, axis=1)

        macro_prec = precision_score(y_true, y_pred, average="macro", zero_division=0)
        macro_rec  = recall_score(y_true, y_pred, average="macro", zero_division=0)
        macro_f1   = f1_score(y_true, y_pred, average="macro", zero_division=0)
        macro_jacc = jaccard_score(y_true, y_pred, average="macro")
        mcc        = matthews_corrcoef(y_true, y_pred)

        labels = np.arange(len(self.class_names))
        cm = confusion_matrix(y_true, y_pred, labels=labels)
        sens = np.diag(cm) / np.maximum(1, cm.sum(axis=1))
        total = cm.sum(); row_sum = cm.sum(axis=1); col_sum = cm.sum(axis=0)
        tn = total - (row_sum + col_sum - np.diag(cm))
        fp = col_sum - np.diag(cm)
        spec = tn / np.maximum(1, tn + fp)
        macro_spec = np.nanmean(spec)

        try:
            macro_auroc = roc_auc_score(y_true, y_prob, multi_class="ovr", average="macro")
        except Exception:
            macro_auroc = np.nan

        row = {
            "epoch": int(epoch+1),
            "lr": self._current_lr(),
            "val_loss": float(logs.get("val_loss", np.nan)),
            "val_accuracy": float(logs.get("val_accuracy", np.nan)),
            "macro_precision": float(macro_prec),
            "macro_recall": float(macro_rec),
            "macro_specificity": float(macro_spec),
            "macro_f1": float(macro_f1),
            "macro_jaccard": float(macro_jacc),
            "mcc": float(mcc),
            "macro_auroc": float(macro_auroc),
            "sens_dr": float(sens[self.idx_dr]) if self.idx_dr is not None else np.nan,
            "spec_dr": float(spec[self.idx_dr]) if self.idx_dr is not None else np.nan,
            "sens_glaucoma": float(sens[self.idx_gl]) if self.idx_gl is not None else np.nan,
            "spec_glaucoma": float(spec[self.idx_gl]) if self.idx_gl is not None else np.nan,
        }
        with open(self.csv_path, "a", newline="") as f:
            csv.DictWriter(f, fieldnames=row.keys()).writerow(row)
        print(
            f"[epoch {row['epoch']}] val_acc={row['val_accuracy']:.4f} "
            f"F1(macro)={row['macro_f1']:.4f} Sens(macro)={row['macro_recall']:.4f} "
            f"Spec(macro)={row['macro_specificity']:.4f} MCC={row['mcc']:.4f} "
            f"AUROC(macro)={row['macro_auroc']:.4f}"
        )

callbacks = [
    keras.callbacks.ModelCheckpoint("efnb3_best.keras", monitor="val_accuracy",
                                    save_best_only=True, save_weights_only=False),
    keras.callbacks.ModelCheckpoint("efnb3_best.h5", monitor="val_accuracy",
                                    save_best_only=True, save_weights_only=False),
    keras.callbacks.ModelCheckpoint("ckpts/latest.weights.h5", monitor="val_accuracy",
                                    save_best_only=False, save_weights_only=True),
    keras.callbacks.BackupAndRestore(backup_dir="backup"),
    csv_logger,
    MedMetricsCSV(val_ds, classes, "logs/val_medical_metrics.csv")
]

# ----------------------------
# Train (end-to-end, no warm-up/freeze)
# ----------------------------
model.compile(optimizer=optimizer, loss=loss_fn, metrics=["accuracy"])

if TRAIN:
    history = model.fit(
        train_ds, validation_data=val_ds,
        epochs=TOTAL_EPOCHS, callbacks=callbacks, verbose=1
    )
else:
    # load best model (from prior run) to evaluate & Grad-CAM
    if os.path.exists("efnb3_best.keras"):
        model = keras.models.load_model("efnb3_best.keras", compile=False)
        print("Loaded saved model: efnb3_best.keras")
    else:
        print("No saved model found; set TRAIN=True to train.")

# ----------------------------
# Plots (same as yours)
# ----------------------------
def plot_history(h):
    if not TRAIN: return
    fig, ax = plt.subplots(1,2, figsize=(12,4))
    ax[0].plot(h.history.get("accuracy", [])); ax[0].plot(h.history.get("val_accuracy", []))
    ax[0].set_title("Accuracy"); ax[0].legend(["train","val"])
    ax[1].plot(h.history.get("loss", [])); ax[1].plot(h.history.get("val_loss", []))
    ax[1].set_title("Loss"); ax[1].legend(["train","val"])
    plt.tight_layout(); plt.show()

if TRAIN:
    plot_history(history)

# ----------------------------
# Final evaluation on test set (same metrics as yours)
# ----------------------------
y_true, y_prob = [], []
for bx, by in test_ds:
    y_true.append(np.argmax(by.numpy(), axis=1))
    y_prob.append(model.predict(bx, verbose=0))
y_true = np.concatenate(y_true); y_prob = np.concatenate(y_prob)
y_pred = np.argmax(y_prob, axis=1)

test_acc = (y_true == y_pred).mean()
print(f"\nTEST Accuracy: {test_acc*100:.2f}%\n")
print("TEST Classification Report:")
print(classification_report(y_true, y_pred, target_names=display_names, digits=4))

cm = confusion_matrix(y_true, y_pred, labels=np.arange(num_classes))
print("\nTEST Confusion Matrix:\n", cm)

mcc            = matthews_corrcoef(y_true, y_pred)
dice_macro     = f1_score(y_true, y_pred, average="macro")
dice_weighted  = f1_score(y_true, y_pred, average="weighted")
jaccard_macro  = jaccard_score(y_true, y_pred, average="macro")
jaccard_weighted = jaccard_score(y_true, y_pred, average="weighted")
try:
    macro_auroc_test = roc_auc_score(y_true, y_prob, multi_class="ovr", average="macro")
except Exception:
    macro_auroc_test = np.nan

print(f"\nMCC: {mcc:.4f}")
print(f"Dice (macro): {dice_macro:.4f}   Dice (weighted): {dice_weighted:.4f}")
print(f"Jaccard (macro): {jaccard_macro:.4f}   Jaccard (weighted): {jaccard_weighted:.4f}")
print(f"AUROC (macro): {macro_auroc_test:.4f}")

def plot_cm(cm, labels, title="Confusion Matrix"):
    fig, ax = plt.subplots(figsize=(6,5))
    im = ax.imshow(cm, cmap="Blues")
    ax.set_xticks(range(len(labels))); ax.set_yticks(range(len(labels)))
    ax.set_xticklabels(labels, rotation=45, ha="right"); ax.set_yticklabels(labels)
    for (i,j), v in np.ndenumerate(cm):
        ax.text(j, i, str(v), ha="center", va="center")
    ax.set_xlabel("Predicted"); ax.set_ylabel("True"); ax.set_title(title)
    plt.colorbar(im); plt.tight_layout(); plt.show()
plot_cm(cm, display_names)

# ----------------------------
# Bootstrap 95% CIs (same as yours)
# ----------------------------
def bootstrap_ci(metric_fn, y_true, y_pred, n=1000, seed=SEED):
    rng = np.random.default_rng(seed)
    idx = np.arange(len(y_true)); vals = []
    for _ in range(n):
        b = rng.choice(idx, size=len(idx), replace=True)
        vals.append(metric_fn(y_true[b], y_pred[b]))
    vals = np.array(vals)
    lo, hi = np.percentile(vals, [2.5, 97.5])
    return float(vals.mean()), (float(lo), float(hi))

acc_mean, acc_ci = bootstrap_ci(lambda yt, yp: (yt == yp).mean(), y_true, y_pred)
f1_mean, f1_ci   = bootstrap_ci(lambda yt, yp: f1_score(yt, yp, average="macro"), y_true, y_pred)
mcc_mean, mcc_ci = bootstrap_ci(lambda yt, yp: matthews_corrcoef(yt, yp), y_true, y_pred)

print("\nBootstrapped 95% CI:")
print(f" Accuracy mean={acc_mean:.4f}  CI=({acc_ci[0]:.4f}, {acc_ci[1]:.4f})")
print(f" F1-macro mean={f1_mean:.4f}  CI=({f1_ci[0]:.4f}, {f1_ci[1]:.4f})")
print(f" MCC mean={mcc_mean:.4f}      CI=({mcc_ci[0]:.4f}, {mcc_ci[1]:.4f})")

# ----------------------------
# Grad-CAM (robust; uses 'top_activation' if available)
# ----------------------------
from tensorflow.keras import Model

# find EfficientNet submodel
eff = model.get_layer("efficientnetb3")
# prefer the activated top; else last conv
layer_for_cam = None
for name in ["top_activation", "top_conv"]:
    try:
        layer_for_cam = eff.get_layer(name)
        print("Grad-CAM base activation:", name)
        break
    except Exception:
        continue
if layer_for_cam is None:
    # fallback: last Conv2D
    layer_for_cam = next(l for l in reversed(eff.layers) if isinstance(l, layers.Conv2D))
    print("Grad-CAM base activation:", layer_for_cam.name)

def make_gradcam(img_tensor, class_idx=None):
    conv_out = layer_for_cam.output
    cam_model = Model(inputs=model.inputs, outputs=[conv_out, model.output])
    with tf.GradientTape() as tape:
        convs, preds = cam_model(img_tensor, training=False)
        if class_idx is None:
            class_idx = tf.argmax(preds[0])
        loss = preds[:, class_idx]
    grads = tape.gradient(loss, convs)
    weights = tf.reduce_mean(grads, axis=(0,1,2))
    convs = convs[0]
    heat  = tf.reduce_sum(convs * weights, axis=-1)
    heat  = tf.maximum(heat, 0) / (tf.reduce_max(heat) + 1e-9)
    return heat.numpy(), int(class_idx.numpy()), preds.numpy()[0]

# choose a sample image
sample_path = test_df.sample(1, random_state=SEED)["filepath"].iloc[0]
raw_bytes   = tf.io.read_file(sample_path)
raw_img     = tf.image.decode_image(raw_bytes, channels=3, expand_animations=False)
raw_float   = tf.image.convert_image_dtype(raw_img, tf.float32)        # [0,1], original size
h0, w0      = raw_float.shape[:2]

x = tf.image.resize(raw_float, (IMG_SIZE, IMG_SIZE))[None, ...]
heat, cls_idx, probs = make_gradcam(x)

# upsample heatmap to original size and overlay
hm = cv2.resize(heat, (w0, h0))
hm = (255 * hm).astype("uint8")
hm = cv2.applyColorMap(hm, cv2.COLORMAP_JET)[:, :, ::-1] / 255.0
overlay = 0.35*hm + 0.65*raw_float.numpy()

pred_name = display_names[cls_idx]
print(f"\nUsing image: {sample_path}")
print(f"Predicted: {pred_name} (idx={cls_idx})")
print("Probabilities:", np.round(probs, 4))

plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.imshow(raw_float.numpy()); plt.axis("off"); plt.title("Original")
plt.subplot(1,3,2); plt.imshow(overlay); plt.axis("off"); plt.title(f"Grad-CAM → {pred_name}")
plt.subplot(1,3,3); plt.bar(range(num_classes), probs); plt.xticks(range(num_classes), display_names, rotation=45); plt.ylim(0,1.0); plt.title("Class probabilities")
plt.tight_layout(); plt.show()

# save overlay next to outputs
base_name = pathlib.Path(sample_path).name.rsplit(".",1)[0]
out_png = f"{base_name}_gradcam.png"
cv2.imwrite(out_png, (overlay[..., ::-1] * 255).astype("uint8"))
print("Saved overlay to:", out_png)

# ----------------------------
# Save final artifacts (same names)
# ----------------------------
try:
    model.save("efnb3_final.keras")
    model.save("efnb3_final.h5")
except Exception as e:
    print("Final HDF5 save failed; saving weights-only fallback:", e)
    model.save_weights("efnb3_final.weights.h5")

print("\nArtifacts written:")
print(" - Best models:  efnb3_best.keras , efnb3_best.h5")
print(" - Last models:  efnb3_final.keras , efnb3_final.h5 (or .weights.h5)")
print(" - Latest ckpt:  ckpts/latest.weights.h5")
print(" - Splits:       splits/train.csv , splits/val.csv , splits/test.csv")
print(" - Logs:         logs/train_history.csv , logs/val_medical_metrics.csv")
```

```python
# =========================================================
# End-to-end (paper-aligned) training + robust Grad-CAM (single cell)
# =========================================================

import os, glob, math, random, pathlib, csv, warnings, sys
warnings.filterwarnings("ignore")

# ----------------------------
# Reproducibility
# ----------------------------
SEED = 1234
os.environ["PYTHONHASHSEED"] = str(SEED)
random.seed(SEED)
import numpy as np
np.random.seed(SEED)

import tensorflow as tf
tf.random.set_seed(SEED)
os.environ["TF_DETERMINISTIC_OPS"] = "1"
os.environ["TF_CUDNN_DETERMINISTIC"] = "1"

import pandas as pd
import matplotlib.pyplot as plt
import cv2

from tensorflow import keras
from tensorflow.keras import layers, Model
from tensorflow.keras.applications import EfficientNetB3
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    classification_report, confusion_matrix, matthews_corrcoef,
    f1_score, jaccard_score, precision_score, recall_score, roc_auc_score
)

# =========================================================
# Config (paper-aligned + robust Grad-CAM)
# =========================================================
IMG_SIZE      = 300
BATCH_SIZE    = 32
TOTAL_EPOCHS  = 70
INIT_LR       = 1e-4
TRAIN         = True       # set False to skip training and just evaluate + Grad-CAM

# Optional: force a Grad-CAM image (leave "" to auto-pick from test set / dataset)
IMG_PATH = ""  # e.g. "/kaggle/input/eye-diseases-classification/dataset/diabetic_retinopathy/1297_right.jpeg"

# =========================================================
# Dataset path detection (tolerant to underscores/spaces)
# =========================================================
from pathlib import Path
from collections import deque

ROOT_CANDIDATES = [
    "/kaggle/input/eye-diseases-classification",
    "/kaggle/input/eye-diseases-classification/dataset",
    "/kaggle/input"
]
def _norm(s: str) -> str: return s.lower().replace("_"," ").strip()
EXPECTED = {"cataract","diabetic retinopathy","glaucoma","normal"}

def find_data_dir_recursive(roots, max_depth=4):
    for root in roots:
        rootp = Path(root)
        if not rootp.exists(): continue
        dq = deque([(rootp, 0)])
        while dq:
            cur, depth = dq.popleft()
            if depth > max_depth: continue
            if cur.is_dir():
                kids = [c.name for c in cur.iterdir() if c.is_dir()]
                if {_norm(k) for k in kids} >= EXPECTED:
                    return str(cur), sorted(kids)
                for child in cur.iterdir():
                    if child.is_dir(): dq.append((child, depth+1))
    raise FileNotFoundError(f"Could not find class folders {EXPECTED} under {roots}")

DATA_DIR, classes = find_data_dir_recursive(ROOT_CANDIDATES, max_depth=4)
print("Using DATA_DIR:", DATA_DIR)
print("Found classes:", classes)

display_names = [c.replace("_"," ") for c in classes]
class_to_id   = {c:i for i,c in enumerate(classes)}
id_to_class   = {i:c for c,i in class_to_id.items()}
num_classes   = len(classes)
print("Label mapping:", class_to_id)

# =========================================================
# List files + 70/20/10 split
# =========================================================
def list_images(root):
    fps, labs = [], []
    for cls in classes:
        for fp in glob.glob(os.path.join(root, cls, "*")):
            if fp.lower().endswith((".jpg",".jpeg",".png",".bmp",".tif",".tiff")):
                fps.append(fp); labs.append(cls)
    return pd.DataFrame({"filepath": fps, "label": labs})

all_df = list_images(DATA_DIR)
print("Total images:", len(all_df))
print(all_df.groupby("label").size())

y_all = all_df["label"].map(class_to_id).values
train_df, temp_df = train_test_split(all_df, test_size=0.30, random_state=SEED, stratify=y_all)
y_temp = temp_df["label"].map(class_to_id).values
val_df, test_df = train_test_split(temp_df, test_size=1/3, random_state=SEED, stratify=y_temp)

print("Split sizes -> Train:", len(train_df), " Val:", len(val_df), " Test:", len(test_df))
print("Train dist:\n", train_df.groupby("label").size())
print("Val   dist:\n", val_df.groupby("label").size())
print("Test  dist:\n", test_df.groupby("label").size())

os.makedirs("splits", exist_ok=True)
train_df.to_csv("splits/train.csv", index=False)
val_df.to_csv("splits/val.csv", index=False)
test_df.to_csv("splits/test.csv", index=False)

# =========================================================
# TF Datasets (paper: conservative aug → horizontal flip)
# =========================================================
AUTOTUNE = tf.data.AUTOTUNE
augment = keras.Sequential([layers.RandomFlip("horizontal")], name="augmentation")

def load_image(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=3, expand_animations=False)
    img = tf.image.convert_image_dtype(img, tf.float32)  # [0,1]
    img = tf.image.resize(img, (IMG_SIZE, IMG_SIZE), antialias=True)
    return img, tf.one_hot(label, num_classes)

def make_ds(df, training=False):
    paths = df["filepath"].values
    y = df["label"].map(class_to_id).values
    ds = tf.data.Dataset.from_tensor_slices((paths, y))
    if training:
        ds = ds.shuffle(buffer_size=len(df), seed=SEED, reshuffle_each_iteration=True)
    ds = ds.map(lambda p,lab: load_image(p, lab), num_parallel_calls=AUTOTUNE)
    if training:
        ds = ds.map(lambda x,y: (augment(x, training=True), y), num_parallel_calls=AUTOTUNE)
    return ds.batch(BATCH_SIZE, drop_remainder=False).prefetch(AUTOTUNE)

train_ds = make_ds(train_df, training=True)
val_ds   = make_ds(val_df,   training=False)
test_ds  = make_ds(test_df,  training=False)

# =========================================================
# Model (EffNetB3 + GAP → DO(0.3) → Dense(256, ReLU) → DO(0.3) → Dense(4, softmax))
# (Dense layers use_bias=True; EffNet convs default bias=False)
# =========================================================
def build_paper_aligned_model(input_shape=(IMG_SIZE, IMG_SIZE, 3), num_classes=num_classes):
    try:
        base = EfficientNetB3(include_top=False, weights="imagenet", input_shape=input_shape)
    except Exception:
        print("ImageNet weights unavailable; using random init.")
        base = EfficientNetB3(include_top=False, weights=None, input_shape=input_shape)
    inputs  = layers.Input(shape=input_shape)
    x       = base(inputs)
    x       = layers.GlobalAveragePooling2D(name="gap")(x)
    x       = layers.Dropout(0.3, name="dropout_1")(x)
    x       = layers.Dense(256, activation="relu", use_bias=True, name="fc_256")(x)
    x       = layers.Dropout(0.3, name="dropout_2")(x)
    outputs = layers.Dense(num_classes, activation="softmax", use_bias=True, name="pred")(x)
    return keras.Model(inputs, outputs, name="efnb3_paper_aligned")

keras.backend.clear_session()
model = build_paper_aligned_model()

# =========================================================
# Optimizer / Loss / Schedule (paper: Adam + cosine annealing)
# =========================================================
steps_per_epoch = max(1, math.ceil(len(train_df) / BATCH_SIZE))
lr_schedule = keras.optimizers.schedules.CosineDecayRestarts(
    initial_learning_rate=INIT_LR,
    first_decay_steps=steps_per_epoch * 10,  # ~10-epoch first cycle
    t_mul=2.0, m_mul=1.0, alpha=1e-6
)
optimizer = keras.optimizers.Adam(learning_rate=lr_schedule)
loss_fn   = keras.losses.CategoricalCrossentropy()

# =========================================================
# Callbacks (best/last saves, logs, and per-epoch medical metrics)
# =========================================================
os.makedirs("logs", exist_ok=True)
os.makedirs("ckpts", exist_ok=True)
os.makedirs("backup", exist_ok=True)

csv_logger = keras.callbacks.CSVLogger("logs/train_history.csv", separator=",", append=True)

class MedMetricsCSV(keras.callbacks.Callback):
    def __init__(self, val_ds, class_names, csv_path="logs/val_medical_metrics.csv"):
        super().__init__()
        self.val_ds, self.class_names = val_ds, list(class_names)
        self.csv_path = csv_path
        norm = lambda s: s.lower().replace("_"," ").strip()
        name_to_idx = {norm(c): i for i, c in enumerate(self.class_names)}
        self.idx_dr = name_to_idx.get("diabetic retinopathy", None)
        self.idx_gl = name_to_idx.get("glaucoma", None)
        if not os.path.exists(self.csv_path):
            with open(self.csv_path, "w", newline="") as f:
                cols = [
                    "epoch","lr","val_loss","val_accuracy",
                    "macro_precision","macro_recall","macro_specificity",
                    "macro_f1","macro_jaccard","mcc","macro_auroc",
                    "sens_dr","spec_dr","sens_glaucoma","spec_glaucoma"
                ]
                csv.writer(f).writerow(cols)
    def _current_lr(self):
        try:
            lr = self.model.optimizer.learning_rate
            if isinstance(lr, tf.keras.optimizers.schedules.LearningRateSchedule):
                return float(lr(self.model.optimizer.iterations).numpy())
            return float(tf.keras.backend.get_value(lr))
        except Exception:
            return float("nan")
    def on_epoch_end(self, epoch, logs=None):
        logs = logs or {}
        y_true, y_prob = [], []
        for bx, by in self.val_ds:
            y_true.append(np.argmax(by.numpy(), axis=1))
            y_prob.append(self.model.predict(bx, verbose=0))
        y_true = np.concatenate(y_true); y_prob = np.concatenate(y_prob)
        y_pred = np.argmax(y_prob, axis=1)

        macro_prec = precision_score(y_true, y_pred, average="macro", zero_division=0)
        macro_rec  = recall_score(y_true, y_pred, average="macro", zero_division=0)
        macro_f1   = f1_score(y_true, y_pred, average="macro", zero_division=0)
        macro_jacc = jaccard_score(y_true, y_pred, average="macro")
        mcc        = matthews_corrcoef(y_true, y_pred)

        labels = np.arange(len(self.class_names))
        cm = confusion_matrix(y_true, y_pred, labels=labels)
        sens = np.diag(cm) / np.maximum(1, cm.sum(axis=1))
        total = cm.sum(); row_sum = cm.sum(axis=1); col_sum = cm.sum(axis=0)
        tn = total - (row_sum + col_sum - np.diag(cm))
        fp = col_sum - np.diag(cm)
        spec = tn / np.maximum(1, tn + fp)
        macro_spec = np.nanmean(spec)

        try:
            macro_auroc = roc_auc_score(y_true, y_prob, multi_class="ovr", average="macro")
        except Exception:
            macro_auroc = np.nan

        row = {
            "epoch": int(epoch+1),
            "lr": self._current_lr(),
            "val_loss": float(logs.get("val_loss", np.nan)),
            "val_accuracy": float(logs.get("val_accuracy", np.nan)),
            "macro_precision": float(macro_prec),
            "macro_recall": float(macro_rec),
            "macro_specificity": float(macro_spec),
            "macro_f1": float(macro_f1),
            "macro_jaccard": float(macro_jacc),
            "mcc": float(mcc),
            "macro_auroc": float(macro_auroc),
            "sens_dr": float(sens[self.idx_dr]) if self.idx_dr is not None else np.nan,
            "spec_dr": float(spec[self.idx_dr]) if self.idx_dr is not None else np.nan,
            "sens_glaucoma": float(sens[self.idx_gl]) if self.idx_gl is not None else np.nan,
            "spec_glaucoma": float(spec[self.idx_gl]) if self.idx_gl is not None else np.nan,
        }
        with open(self.csv_path, "a", newline="") as f:
            csv.DictWriter(f, fieldnames=row.keys()).writerow(row)
        print(
            f"[epoch {row['epoch']}] val_acc={row['val_accuracy']:.4f} "
            f"F1(macro)={row['macro_f1']:.4f} Sens(macro)={row['macro_recall']:.4f} "
            f"Spec(macro)={row['macro_specificity']:.4f} MCC={row['mcc']:.4f} "
            f"AUROC(macro)={row['macro_auroc']:.4f}"
        )

callbacks = [
    keras.callbacks.ModelCheckpoint("efnb3_best.keras", monitor="val_accuracy",
                                    save_best_only=True, save_weights_only=False),
    keras.callbacks.ModelCheckpoint("efnb3_best.h5", monitor="val_accuracy",
                                    save_best_only=True, save_weights_only=False),
    keras.callbacks.ModelCheckpoint("ckpts/latest.weights.h5", monitor="val_accuracy",
                                    save_best_only=False, save_weights_only=True),
    keras.callbacks.BackupAndRestore(backup_dir="backup"),
    csv_logger,
    MedMetricsCSV(val_ds, classes, "logs/val_medical_metrics.csv")
]

# =========================================================
# Train (or load best)
# =========================================================
model.compile(optimizer=optimizer, loss=loss_fn, metrics=["accuracy"])

if TRAIN:
    history = model.fit(
        train_ds, validation_data=val_ds,
        epochs=TOTAL_EPOCHS, callbacks=callbacks, verbose=1
    )
else:
    def _find_first(patterns, roots):
        for root in roots:
            for pat in patterns:
                hits = glob.glob(os.path.join(root, "**", pat), recursive=True)
                if hits:
                    # prefer /kaggle/working over /kaggle/input
                    hits = sorted(hits, key=lambda p: ("/kaggle/working" not in p, len(p)))
                    return hits[0]
        return None
    best_path = _find_first(
        ["efnb3_best.keras", "efnb3_final.keras", "efnb3_best.h5", "efnb3_final.h5"],
        ["/kaggle/working", "/kaggle/input"]
    )
    if best_path:
        print(f"Loading saved model: {best_path}")
        model = keras.models.load_model(best_path, compile=False)
    else:
        print("No saved model found; set TRAIN=True to train."); raise SystemExit(0)

# =========================================================
# Plots (if we trained now)
# =========================================================
def plot_history(h):
    fig, ax = plt.subplots(1,2, figsize=(12,4))
    ax[0].plot(h.history.get("accuracy", [])); ax[0].plot(h.history.get("val_accuracy", []))
    ax[0].set_title("Accuracy"); ax[0].legend(["train","val"])
    ax[1].plot(h.history.get("loss", [])); ax[1].plot(h.history.get("val_loss", []))
    ax[1].set_title("Loss"); ax[1].legend(["train","val"])
    plt.tight_layout(); plt.show()

if TRAIN:
    plot_history(history)

# =========================================================
# Final evaluation on the held-out test set
# =========================================================
y_true, y_prob = [], []
for bx, by in test_ds:
    y_true.append(np.argmax(by.numpy(), axis=1))
    y_prob.append(model.predict(bx, verbose=0))
y_true = np.concatenate(y_true); y_prob = np.concatenate(y_prob)
y_pred = np.argmax(y_prob, axis=1)

test_acc = (y_true == y_pred).mean()
print(f"\nTEST Accuracy: {test_acc*100:.2f}%\n")
print("TEST Classification Report:")
print(classification_report(y_true, y_pred, target_names=display_names, digits=4))

cm = confusion_matrix(y_true, y_pred, labels=np.arange(num_classes))
print("\nTEST Confusion Matrix:\n", cm)

mcc             = matthews_corrcoef(y_true, y_pred)
dice_macro      = f1_score(y_true, y_pred, average="macro")
dice_weighted   = f1_score(y_true, y_pred, average="weighted")
jaccard_macro   = jaccard_score(y_true, y_pred, average="macro")
jaccard_weighted= jaccard_score(y_true, y_pred, average="weighted")
try:
    macro_auroc_test = roc_auc_score(y_true, y_prob, multi_class="ovr", average="macro")
except Exception:
    macro_auroc_test = np.nan

print(f"\nMCC: {mcc:.4f}")
print(f"Dice (macro): {dice_macro:.4f}   Dice (weighted): {dice_weighted:.4f}")
print(f"Jaccard (macro): {jaccard_macro:.4f}   Jaccard (weighted): {jaccard_weighted:.4f}")
print(f"AUROC (macro): {macro_auroc_test:.4f}")

def plot_cm(cm, labels, title="Confusion Matrix"):
    fig, ax = plt.subplots(figsize=(6,5))
    im = ax.imshow(cm, cmap="Blues")
    ax.set_xticks(range(len(labels))); ax.set_yticks(range(len(labels)))
    ax.set_xticklabels(labels, rotation=45, ha="right"); ax.set_yticklabels(labels)
    for (i,j), v in np.ndenumerate(cm):
        ax.text(j, i, str(v), ha="center", va="center")
    ax.set_xlabel("Predicted"); ax.set_ylabel("True"); ax.set_title(title)
    plt.colorbar(im); plt.tight_layout(); plt.show()
plot_cm(cm, display_names)

# =========================================================
# Bootstrap 95% CIs
# =========================================================
def bootstrap_ci(metric_fn, yt, yp, n=1000, seed=SEED):
    rng = np.random.default_rng(seed)
    idx = np.arange(len(yt)); vals = []
    for _ in range(n):
        b = rng.choice(idx, size=len(idx), replace=True)
        vals.append(metric_fn(yt[b], yp[b]))
    vals = np.array(vals)
    lo, hi = np.percentile(vals, [2.5, 97.5])
    return float(vals.mean()), (float(lo), float(hi))

acc_mean, acc_ci = bootstrap_ci(lambda a,b:(a==b).mean(), y_true, y_pred)
f1_mean, f1_ci   = bootstrap_ci(lambda a,b:f1_score(a,b,average="macro"), y_true, y_pred)
mcc_mean, mcc_ci = bootstrap_ci(lambda a,b:matthews_corrcoef(a,b), y_true, y_pred)

print("\nBootstrapped 95% CI:")
print(f" Accuracy mean={acc_mean:.4f}  CI=({acc_ci[0]:.4f}, {acc_ci[1]:.4f})")
print(f" F1-macro mean={f1_mean:.4f}  CI=({f1_ci[0]:.4f}, {f1_ci[1]:.4f})")
print(f" MCC mean={mcc_mean:.4f}      CI=({mcc_ci[0]:.4f}, {mcc_ci[1]:.4f})")

# =========================================================
# Robust Grad-CAM (eager; avoids Functional.call KeyError)
# =========================================================
# 1) find the correct last conv layer (prefer top_activation → top_conv)
def find_last_conv_for_cam(m):
    eff = None
    # try direct submodel first
    try:
        eff = m.get_layer("efficientnetb3")
    except Exception:
        for lyr in m.layers:
            if hasattr(lyr, "layers"):
                try:
                    _ = lyr.get_layer("top_conv")
                    eff = lyr; break
                except Exception:
                    pass
    # preferred layers
    if eff is not None:
        for name in ["top_activation", "top_conv"]:
            try:
                return eff.get_layer(name)
            except Exception:
                pass
    # fallback: last Conv2D anywhere (also search nested)
    convs = [l for l in m.layers if isinstance(l, layers.Conv2D)]
    for lyr in m.layers:
        if hasattr(lyr, "layers"):
            convs += [l for l in lyr.layers if isinstance(l, layers.Conv2D)]
    return convs[-1] if convs else None

last_conv = find_last_conv_for_cam(model)
if last_conv is None:
    print("❌ Could not locate a convolution layer for Grad-CAM. Skipping CAM.")
else:
    print("Grad-CAM using layer:", last_conv.name)

    # 2) choose an image
    if IMG_PATH == "":
        # Prefer an image from test set
        if len(test_df):
            IMG_PATH = test_df.sample(1, random_state=SEED)["filepath"].iloc[0]
        else:
            # fallback: first image from dataset
            for cls in classes:
                hits = glob.glob(os.path.join(DATA_DIR, cls, "*"))
                if hits:
                    IMG_PATH = hits[0]; break
    if not os.path.exists(IMG_PATH):
        print("❌ Could not find an image for Grad-CAM. Set IMG_PATH to a valid file.")
    else:
        print(f"Using image for Grad-CAM: {IMG_PATH}")

        # read + prep
        raw = tf.io.read_file(IMG_PATH)
        raw = tf.image.decode_image(raw, channels=3, expand_animations=False)
        rawf = tf.image.convert_image_dtype(raw, tf.float32)
        x = tf.image.resize(rawf, (IMG_SIZE, IMG_SIZE), antialias=True)[None, ...]

        # 3) CAM model & eager tape to avoid KeyError
        old_flag = tf.config.functions_run_eagerly()
        tf.config.run_functions_eagerly(True)
        try:
            cam_model = Model(model.inputs, [last_conv.output, model.output])
            _ = cam_model(x, training=False)  # warmup run

            with tf.GradientTape() as tape:
                convs, preds = cam_model(x, training=False)
                class_idx = tf.argmax(preds[0])
                loss = preds[:, class_idx]
            grads = tape.gradient(loss, convs)

            if grads is None:
                # rare case; try watching convs explicitly
                with tf.GradientTape() as tape2:
                    tape2.watch(convs)
                    convs2, preds2 = cam_model(x, training=False)
                    class_idx2 = tf.argmax(preds2[0])
                    loss2 = preds2[:, class_idx2]
                grads = tape2.gradient(loss2, convs2)
                convs = convs2; preds = preds2; class_idx = class_idx2

            # weights = spatial GAP over gradients
            weights = tf.reduce_mean(grads, axis=(0,1,2))
            cam = tf.reduce_sum(weights * convs[0], axis=-1)
            cam = tf.maximum(cam, 0) / (tf.reduce_max(cam) + 1e-9)
            cam = cam.numpy()

            # overlay at original size
            h, w = int(rawf.shape[0]), int(rawf.shape[1])
            cam_up = cv2.resize(cam, (w, h), interpolation=cv2.INTER_LINEAR)
            heat = (255 * cam_up).astype("uint8")
            heat_rgb = cv2.applyColorMap(heat, cv2.COLORMAP_JET)[:, :, ::-1] / 255.0
            overlay = (0.35 * heat_rgb + 0.65 * rawf.numpy()).clip(0, 1)

            probs = preds.numpy()[0]
            pred_idx = int(class_idx.numpy())
            pred_name = display_names[pred_idx] if pred_idx < len(display_names) else f"class_{pred_idx}"
            print(f"\nPredicted: {pred_name} (idx={pred_idx})")
            print("Probabilities:", np.round(probs, 4))

            plt.figure(figsize=(12,4))
            plt.subplot(1,3,1); plt.imshow(rawf.numpy()); plt.axis("off"); plt.title("Original")
            plt.subplot(1,3,2); plt.imshow(overlay); plt.axis("off"); plt.title(f"Grad-CAM → {pred_name}")
            plt.subplot(1,3,3); plt.bar(range(len(probs)), probs)
            plt.xticks(range(len(probs)), display_names, rotation=45, ha="right"); plt.ylim(0,1.0)
            plt.title("Class probabilities"); plt.tight_layout(); plt.show()

            out_png = f"{pathlib.Path(IMG_PATH).stem}_gradcam.png"
            cv2.imwrite(out_png, (overlay[..., ::-1] * 255).astype("uint8"))
            print("Saved overlay to:", out_png)

        finally:
            tf.config.run_functions_eagerly(old_flag)

# =========================================================
# Save final artifacts (best already saved by callbacks)
# =========================================================
try:
    model.save("efnb3_final.keras")
    model.save("efnb3_final.h5")
except Exception as e:
    print("Final HDF5 save failed; saving weights-only fallback:", e)
    model.save_weights("efnb3_final.weights.h5")

print("\nArtifacts written:")
print(" - Best models:  efnb3_best.keras , efnb3_best.h5")
print(" - Last models:  efnb3_final.keras , efnb3_final.h5 (or .weights.h5)")
print(" - Latest ckpt:  ckpts/latest.weights.h5")
print(" - Splits:       splits/train.csv , splits/val.csv , splits/test.csv")
print(" - Logs:         logs/train_history.csv , logs/val_medical_metrics.csv")
```

```python
# ===============================
# Robust Grad-CAM (no retrain needed)
# ===============================
import os, glob, pathlib, numpy as np, cv2, tensorflow as tf, matplotlib.pyplot as plt
from tensorflow.keras.models import load_model, Model
from tensorflow.keras import layers

# ---- config
MODEL_PATH = "/kaggle/working/efnb3_best.keras"   # change if your best file is elsewhere
IMG_PATH   = "/kaggle/input/eye-diseases-classification/dataset/diabetic_retinopathy/1297_right.jpeg"  # or "" to auto-pick
IMG_SIZE   = 300
CLASS_NAMES = ["cataract","diabetic retinopathy","glaucoma","normal"]

# ---- tidy up & load fresh
tf.keras.backend.clear_session()
tf.config.optimizer.set_jit(False)                 # avoid XLA weirdness
tf.config.run_functions_eagerly(True)              # force eager for this cell

assert os.path.exists(MODEL_PATH), f"Model not found: {MODEL_PATH}"
model = load_model(MODEL_PATH, compile=False)
print("Loaded model:", MODEL_PATH)

# ---- choose image
def auto_pick():
    roots = ["/kaggle/input/eye-diseases-classification/dataset", "/kaggle/input"]
    exts = (".jpg",".jpeg",".png",".bmp",".tif",".tiff")
    for root in roots:
        if not os.path.isdir(root): continue
        hits = [p for p in glob.glob(os.path.join(root, "**", "*"), recursive=True) if p.lower().endswith(exts)]
        if hits: return sorted(hits)[0]
    return None

if not IMG_PATH:
    IMG_PATH = auto_pick()
assert IMG_PATH and os.path.exists(IMG_PATH), f"Image not found: {IMG_PATH}"
print("Using image:", IMG_PATH)

# ---- prep image
raw = tf.io.read_file(IMG_PATH)
raw = tf.image.decode_image(raw, channels=3, expand_animations=False)
rawf = tf.image.convert_image_dtype(raw, tf.float32)
x = tf.image.resize(rawf, (IMG_SIZE, IMG_SIZE), antialias=True)[None, ...]

# ---- find EfficientNet submodel and candidate conv layers
def find_eff_submodel(m):
    try:
        return m.get_layer("efficientnetb3")
    except Exception:
        for lyr in m.layers:
            if hasattr(lyr, "layers"):
                try:
                    _ = lyr.get_layer("top_conv")
                    return lyr
                except Exception:
                    pass
    return None

eff = find_eff_submodel(model)
if eff is None:
    # fallback: scan whole model for convs
    conv_candidates = [l for l in model.layers if isinstance(l, layers.Conv2D)]
else:
    conv_candidates = [l for l in eff.layers if isinstance(l, layers.Conv2D)]
# prefer very last convs first
conv_candidates = list(reversed(conv_candidates))

# put the usual suspects first if present
priority = ["top_conv", "block7a_project_conv", "block6d_project_conv"]
name_to_layer = {l.name: l for l in conv_candidates}
ordered = [name_to_layer[n] for n in priority if n in name_to_layer]
for l in conv_candidates:
    if l not in ordered:
        ordered.append(l)

# ---- try layers until one works
chosen = None
convs = preds = None
for cand in ordered:
    try:
        grad_model = Model(model.inputs, [cand.output, model.output])
        with tf.GradientTape() as tape:
            convs, preds = grad_model(x, training=False)   # <-- where the KeyError used to happen
            class_idx = tf.argmax(preds[0])
            loss = preds[:, class_idx]
        grads = tape.gradient(loss, convs)
        if grads is not None:
            chosen = cand
            break
    except Exception as e:
        # try next candidate
        continue

if chosen is not None:
    print("Grad-CAM layer:", chosen.name)
    # standard CAM weights (global-average pool of grads)
    weights = tf.reduce_mean(grads, axis=(0,1,2))
    cam = tf.reduce_sum(weights * convs[0], axis=-1)
    cam = tf.maximum(cam, 0) / (tf.reduce_max(cam) + 1e-9)
    cam = cam.numpy()
else:
    # ---------- SALIENCY FALLBACK (no intermediate model needed) ----------
    print("⚠️ All candidate conv layers failed. Falling back to input-gradient saliency.")
    with tf.GradientTape() as tape:
        tape.watch(x)
        preds = model(x, training=False)
        class_idx = tf.argmax(preds[0])
        loss = preds[:, class_idx]
    gx = tape.gradient(loss, x)[0]                 # [H,W,3]
    cam = tf.reduce_max(tf.abs(gx), axis=-1)
    cam = cam / (tf.reduce_max(cam) + 1e-9)
    cam = cam.numpy()

# ---- overlay
h, w = int(rawf.shape[0]), int(rawf.shape[1])
hm = cv2.resize(cam, (w, h), interpolation=cv2.INTER_LINEAR)
heat = (255 * hm).astype("uint8")
heat_rgb = cv2.applyColorMap(heat, cv2.COLORMAP_JET)[:, :, ::-1] / 255.0
overlay = (0.35 * heat_rgb + 0.65 * rawf.numpy()).clip(0, 1)

probs = preds.numpy()[0]
cidx = int(class_idx.numpy()) if hasattr(class_idx, "numpy") else int(class_idx)
label = CLASS_NAMES[cidx] if cidx < len(CLASS_NAMES) else f"class_{cidx}"
print(f"Predicted: {label} (idx={cidx})")
print("Probabilities:", np.round(probs, 4))

plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.imshow(rawf.numpy()); plt.axis("off"); plt.title("Original")
plt.subplot(1,3,2); plt.imshow(overlay); plt.axis("off"); plt.title(("Grad-CAM" if chosen else "Saliency") + f" → {label}")
plt.subplot(1,3,3); plt.bar(range(len(probs)), probs); 
xt = CLASS_NAMES if len(CLASS_NAMES)==len(probs) else [f"c{i}" for i in range(len(probs))]
plt.xticks(range(len(probs)), xt, rotation=45, ha="right"); plt.ylim(0,1.0); plt.title("Class probabilities")
plt.tight_layout(); plt.show()

out_png = f"{pathlib.Path(IMG_PATH).stem}_{'gradcam' if chosen else 'saliency'}.png"
cv2.imwrite(out_png, (overlay[:, :, ::-1] * 255).astype("uint8"))
print("Saved overlay to:", out_png)
```
